```python
#Data saved in csv/excel
import glob, os
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

data_dir="R:\MSS\Research\Projects\Baby_Toolbox\Validation_Data\misc"
os.chdir(data_dir)
```

# Get the data

```python
# get all assessment files
afiles = glob.glob('R:\\MSS\\Research\\Projects\\Baby_Toolbox\\Validation_Data\\misc\\*Assessment Data.csv')

doc =[]
for file in afiles:
    if '2021-11-1' in file and 'Narrow Structure' not in file:
        data = pd.read_csv(file)
        print(file)
        doc.append(data)

# get all reg files 
rdoc =[]
rfiles = glob.glob('R:\\MSS\\Research\\Projects\\Baby_Toolbox\\Validation_Data\\misc\\*Registration Data.csv')
for file in rfiles:
    if '2021-11-1' in file and 'Narrow Structure' not in file:
        data = pd.read_csv(file)
        print(file)
        rdoc.append(data)

adf = pd.concat(doc)
rdf = pd.concat(rdoc)
```

```python
rdf= rdf.drop(dropindex)
```

```python
#sanity check
rdf.groupby("PIN").filter(lambda x: len(x) > 1)
```

```python
#Set two decimal points for display
pd.options.display.float_format = "{:.2f}".format
```

```python
rdf[['Year','Month']] = rdf.Age.str.split(":",expand=True)
```

```python
rdf['Month'] =pd.to_numeric(rdf['Month'])
rdf['Year'] =pd.to_numeric(rdf['Year'])
```

```python
#recode the age into months format
rdf['Age_month'] = rdf['Year']*12 + rdf['Month']
```

```python
rdf= rdf.drop_duplicates()
```

```python
adf= adf.drop_duplicates()
```

```python
adf.shape
```

```python
#Recode age group
rdf['Age_group'] = pd.cut(rdf['Age_month'] , [0,1,2,3,4,5,6,8,10,12,15,18,21,24,27,30,35,43])
```

```python
# Concatenate registration and assessment data files in the /misc folder
df = pd.merge(adf, rdf[['PIN','Assessment Name', 'Age_month','Age_group']], on=['PIN','Assessment Name'], how ='left')
```

```python
#Convert to datetime
df['InstStarted']= pd.to_datetime(df['InstStarted'])
df['InstEnded']= pd.to_datetime(df['InstEnded'])
df['DateCreated']= pd.to_datetime(df['DateCreated'])
```

```python
def count_freq(row):
    dc ={}
    for i in row:
        if i not in dc:
            dc[i] = 0
        else:
            dc[i]+=1
    return dc
```

```python
task4['freq'] =task4['Inst'].apply(count_freq)
```
