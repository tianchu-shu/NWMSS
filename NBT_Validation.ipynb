# 6.       Measure durations

```python
#Convert to datetime
df['InstStarted']= pd.to_datetime(df['InstStarted'])
df['InstEnded']= pd.to_datetime(df['InstEnded'])
df['DateCreated']= pd.to_datetime(df['DateCreated'])
```

```python
#groupby PIN, Inst to calculate the time spent in each inst by each participant
mni =df.groupby(['PIN','Age_group','Inst'])['InstStarted','InstEnded'].first().reset_index()
```

```python
#a. Calculate end time minus start time by measure by PIN
mni['Time']=mni['InstEnded']-mni['InstStarted']
```

```python
mni['Time'].dt.total_seconds()
```

```python
mni['Minute']=round(mni['Time'].dt.total_seconds()/60,2)
```

```python
mni.head()
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('C:/Users/tsb7592/Downloads/task6.xlsx', engine='xlsxwriter')

#b. Summarize (inc. medians) by measure, overall and by age group
mni.groupby(['Inst','Age_group'])['Minute'].describe().to_excel(writer, sheet_name='Summarize_by_inst_age')

mni.groupby(['Inst'])['Minute'].describe().to_excel(writer, sheet_name='Overall_time_by_inst')

#c. Add up the durations within a PIN and summarize total time per PIN (even though they will have a different number of measures) – overall and by age group
mni.groupby(['PIN','Age_group'])['Minute'].sum().to_excel(writer, sheet_name='Overall_duration')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
