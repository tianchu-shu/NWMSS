# Get the data

```python
# Get the files from each city
rfiles = glob.glob('R:\\MSS\\Research\\NIHTB v3.0 Norming\\Riverside Data*\\*Registration Data.csv')

doc =[]
for file in rfiles:
    if 'Narrow Structure' in file:
        pass
    else:
        data = pd.read_csv(file)
        print(file)
        doc.append(data)

rdf = pd.concat(doc)
```
