```python
adults = adults.rename({  '[Q07] - What is your race? American Indian or Alaska Native':'AmericanIndian/AlaskaNative',
       '[Q07] - What is your race? Asian':'QAsian',
       '[Q07] - What is your race? Black or African-American':'QBlack',
       '[Q07] - What is your race? Middle Eastern or North African' :'MiddleEastern',
       '[Q07] - What is your race? Native Hawaiian':'NativeHawaiian',
       '[Q07] - What is your race? Other Pacific Islander' :'OtherPacificIslander',
       '[Q07] - What is your race? White or Caucasian':'QWhite',
        '[Q07] - What is your race? Don`t Know/Prefer not to sayNone of the above':'None'}, axis=1)
```

```python
adults['QOther']  = adults[['AmericanIndian/AlaskaNative',
       'MiddleEastern', 'NativeHawaiian',
       'OtherPacificIslander','None']].sum(axis=1)
```

```python
adults['QOther'].value_counts()
```

```python
def map_score(score):
    #1=No; not of Hispanic; Latinx; or Spanish origin
    if score ==1 :
        return 0
    else:
        return 1

adults["Hispanic"] = adults['[Q06] - Are you Hispanic or Latinx?'].apply(lambda score: map_score(score))
```

```python
adults['sum']  = adults[[ 'QAsian', 'QBlack', 'QWhite', 'QOther']].sum(axis=1)
```

```python
adults['sum'].value_counts()
```

```python
#Sanity check
adults[adults['sum'] == 3][[ 'QAsian', 'QBlack', 'QWhite', 'QOther']]
```

```python
adults['Adult- race/eth'].value_counts()
```

```python
#Hispanic	White	Black	Asian	OtherSingleRace	White/Black	White/Asian	Black/Asian	Multi

adults['Asian'] = adults['sum']* adults['QAsian'] 

def map_score(score):
    if score ==1 :
        return 1
    else:
        return 0

adults['Asian']= adults['Asian'].apply(lambda score: map_score(score))
```

```python
adults['Black'] = adults['sum']* adults['QBlack'] 

adults['Black']= adults['Black'].apply(lambda score: map_score(score))
```

```python
adults['White'] = adults['sum']* adults['QWhite'] 

adults['White']= adults['White'].apply(lambda score: map_score(score))
```

```python
adults['OtherSingleRace'] = adults['sum']* adults['QOther'] 

adults['OtherSingleRace']= adults['OtherSingleRace'].apply(lambda score: map_score(score))
```

```python
adults['OtherSingleRace'].value_counts()
```

```python
adults['White/Asian'] = adults['QAsian']* adults['QWhite'] 
adults['White/Asian'].value_counts()
```

```python
adults['White/Black'] = adults['QBlack']* adults['QWhite'] 
adults['White/Black'].value_counts()
```

```python
adults['Black/Asian'] = adults['QBlack']* adults['QAsian'] 
adults['Black/Asian'].value_counts()
```

```python
def EOQ(sc, ot):
    if sc ==3 :
        return 1
    elif sc ==2 and ot == 1:
        return 1
    else:
        return 0


adults["Multi"] = adults.apply(lambda row: EOQ(row['sum'], row['QOther']), axis=1)
```

```python
#Sanity check
adults[adults['sum'] == 2][[ 'QAsian', 'QBlack', 'QWhite', 'QOther','Multi']]
```

```python
##Hispanic	White	Black	Asian	OtherSingleRace	White/Black	White/Asian	Black/Asian	Multi
adults[['Hispanic','White', 'Black','Asian', 'OtherSingleRace', 'White/Black', 'White/Asian', 'Black/Asian',
       'Multi']]
```

```python
race =['White', 'Black','Asian', 'OtherSingleRace', 'White/Black', 'White/Asian', 'Black/Asian',
       'Multi']
```

```python
for col in race:
    adults[col] = np.where(adults['Hispanic']==1, 0, adults[col])
```

```python
keep = ['Region', 'Hispanic']+race
```
