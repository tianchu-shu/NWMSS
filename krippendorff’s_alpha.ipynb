```python
import sys 

stdoutOrigin=sys.stdout 
sys.stdout = open("krippendorff's_alpha_output.txt", "w")
```

```python
def ka(df, name='TOR'):
    a =simpledorff.calculate_krippendorffs_alpha_for_df(df, experiment_col='Participant ID (video file name)',
                                                 annotator_col='Rater',class_col='Baseline Behavior')
    b = simpledorff.calculate_krippendorffs_alpha_for_df(df, experiment_col='Participant ID (video file name)',
                                                 annotator_col='Rater',   class_col='Intensity')
    print("Baseline Behavior: ", round(a,2))
    print("Intensity: ", round(b,2))
    
    if name == 'TOR':
        c=simpledorff.calculate_krippendorffs_alpha_for_df(df, experiment_col='Participant ID (video file name)',
                                                 annotator_col='Rater',  class_col='Escalation')
        d = simpledorff.calculate_krippendorffs_alpha_for_df(df, experiment_col='Participant ID (video file name)',
                                                 annotator_col='Rater',   class_col='Coping')
        print("Escalation: ", round(c,2))
        print("Coping: ", round(d,2))
        
    else:
        c=simpledorff.calculate_krippendorffs_alpha_for_df(df, experiment_col='Participant ID (video file name)',
                                                 annotator_col='Rater',  class_col='Predominance ')
        print("Predominance: ", round(c,2))
    print('')    
    return
```

```python
data = {'TOR1': df1, 'TOR2': df2,'TOR3': df3,'Bubbles':bb}
```

```python
for name in data:
    
    print(name)
    ka(data[name], name[:3] )
```
