```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('C:/Users/tsb7592/Downloads/Phoenix_compare.xlsx', engine='xlsxwriter')

for folder in glob.glob('C:/Users/tsb7592/Downloads/Phoenix_056or773/*/*.xls*')  :
    if 'docx' in folder:
        pass
    elif '1.5s' in folder:
        pass
    else:
        tdc ={}
        fname = folder.split("\\")[-2]
        print(folder.split("\\")[-1])
        print(fname)
        hc =pd.read_excel(folder)
        col = hc.filter(like='Handcoding_').columns[0]
        hc =hc.loc[hc[col].isin(['right','left','away'])]
        a = hc[col].value_counts().to_dict()
        ap = round(hc[col].value_counts(normalize= True),3).to_dict()

        ad = 'C:/Users/tsb7592/Downloads/Phoenix_videos/'+fname+'_annotation.txt'
        vd = pd.read_csv(ad, header=None)
        vd.columns =['index','TrackName']
        vd[['TrackName']]=vd[['TrackName']].apply(lambda x : x.str.strip())
        b= vd.TrackName.value_counts().to_dict()
        bp = round(vd.TrackName.value_counts(normalize= True),3).to_dict()
        tdc['Handcode'] = a
        tdc['iCatcher'] = b
        tdc['Handcode_percent'] = ap
        tdc['iCatcher_percent'] = bp       
        
        data =pd.DataFrame.from_dict(tdc)
        display(data)
        # Write each dataframe to a different worksheet.
        data.to_excel(writer, sheet_name=fname.split("~")[-1])
        

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```

```python
import cv2
import numpy as np
import glob
import re

def image_convert(folder, tally):
    img_array = []
    imgs =sorted(glob.glob(folder +'/*.jpg'), key=numericalSort)
    for filename in imgs:
        if imgs:
            img = cv2.imread(filename)
            height, width, layers = img.shape
            size = (width,height)
            img_array.append(img)
        else:
            return
 
    name = str(tally)+'_video.mp4'
    #framerate 12 frames per second
    out = cv2.VideoWriter('C:/Users/tsb7592/Downloads/Phoenix_videos/'+name ,cv2.VideoWriter_fourcc(*'DIVX'), 12, size)
 
    for i in range(len(img_array)):
        out.write(img_array[i])
    out.release()
```

```python
numbers = re.compile(r'(\d+)')
                     
def numericalSort(value):
    parts = numbers.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts

tally = 0
for folder in glob.glob('R:/MSS/Research/Projects/Baby_Toolbox/Validation_Data/LookListening_*'):
    if tally%10 == 0:
        print(tally)
    print(folder)
    if tally < 81:
        pass
    else:
        image_convert(folder, tally)
    tally += 1
```
