 # Give N 

```python
data=pd.read_csv('C:/Users/tsb7592/Downloads/Basic_Desc_output.csv')

data['PIN']=data['PIN'].str.upper()
```

```python
def create_sum(df, colist, new, exception = None, allvalid=False):
    
    mni= data[data.ItemID.isin(colist)][['PIN','Assessment Name','ItemID','Score']]
    mni= mni.groupby(['PIN','Assessment Name','ItemID'])['Score'].sum().unstack()
    mni =mni.reset_index()
    if not allvalid:
        mni =mni[mni.notnull().all(1)]
    
    if exception:
        mni = mni[~mni.isin([exception]).any(axis=1)]
    
    mni[new]=mni.sum(axis=1)
    mni = mni.drop(columns = colist)
    
    mni.columns =['ChildPIN','Assessment Name',new]
    
    return mni.drop_duplicates()
```

```python
given =create_sum(data, ['GN_3b','GN_4b',  'GN_2b', 'GN_1b'], 'Give_N')
```

```python
given.head()
```

```python
sc =sc.combine_first(given)
```

```python
sc.shape
```

```python
sc.groupby("ChildPIN").filter(lambda x: len(x) > 2)[keep]
```
