# 4a. Sitting duration: extract the longest duration from ItemID BsitSeg and from Bsit-ReSeg (if administered)

```python
def get_max_duration(df,cola, colb=None):
    if colb:
        seg = df[(df['ItemID'] ==cola)|(df['ItemID'] == colb)].Response.str.split("|",expand=True)
    else:
        seg = df[(df['ItemID'] ==cola)].Response.str.split("|",expand=True)
    slist = []
    for i in range(len(seg.columns)):
        slist.append(seg[i])
        
    seg = pd.concat(slist)
    seg= seg[seg.notnull()].str.split("%",expand=True)
    if colb:
        pref, _ = cola.split('-')
    else:
        _, pref = cola.split('-')
    seg.columns=['_', 'duration']
    seg['duration'] = pd.to_numeric(seg['duration'])

    print(seg['duration'].max())
    seg.columns=['_', pref+'-duration']
    seg.drop('_', axis=1, inplace=True)
    return  seg
```

```python
a =get_max_duration(df, 'Bsit-Reg', 'Bsit-ReSeg')
a
```

# 5. Summarize the five durations (Sitting, Standoverall and by group (number of individuals taking the test, mean duration, sd of durations, minimum, 25th percentile, median, 75th percentile, and maximum durations)
