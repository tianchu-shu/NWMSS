```python
task4.columns=['unique_inst_count', 'Inst']
task4.head()
```

```python
import ast
def count_freq(row):
    dc ={}
    for i in row:
        if i not in dc:
            dc[i] = 0
        else:
            dc[i]+=1
   # convertedDict = ast.literal_eval(dc)

    return dc
```

```python
task4['freq'] =task4['Inst'].apply(count_freq)
```

```python
task4.tail()
```

```python
for index, row in task4.iterrows():
    dictionary = row['freq']
    for key, v in dictionary.items():
        task4.loc[index,key] = v
```

# Basic Descriptives 9-11      Measure durations

```python
#Convert to datetime
df['InstStarted']= pd.to_datetime(df['InstStarted'])
df['InstEnded']= pd.to_datetime(df['InstEnded'])
df['DateCreated']= pd.to_datetime(df['DateCreated'])
```

```python
#groupby PIN, Inst to calculate the time spent in each inst by each participant
mni =df.groupby(['PIN','DYAD','Inst'])['InstStarted','InstEnded'].first().reset_index()
```

```python
#a. Calculate end time minus start time by measure by PIN
mni['Time']=mni['InstEnded']-mni['InstStarted']
```

```python
mni['Time'].dt.total_seconds()
```

```python
mni['Minute']=round(mni['Time'].dt.total_seconds()/60,2)
```

```python
mni.head()
```

```python
mni =pd.merge(mni, rdf, on =['PIN','DYAD'])
```

```python
mni[mni.Age_group.notnull()].shape[0]/mni.shape[0]
```

```python
mni.groupby(['Inst','Age_group'])['Minute'].describe()
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('C:\\Users\\tsb7592\\OneDrive - Northwestern University\\Analysis Progress\\Basic_Desc_task9-11.xlsx', engine='xlsxwriter')

#b. Summarize (inc. medians) by measure, overall and by age group
mni.groupby(['Inst','Age_group'])['Minute'].describe().to_excel(writer, sheet_name='Summarize_by_inst_age')

mni.groupby(['Inst'])['Minute'].describe().to_excel(writer, sheet_name='Overall_time_by_inst')

#c. Add up the durations within a PIN and summarize total time per PIN (even though they will have a different number of measures) – overall and by age group
mni.groupby(['Age_group','PIN'])['Minute'].sum().to_excel(writer, sheet_name='Overall_duration')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
