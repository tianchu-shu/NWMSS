```python
def onetwothree(df):
    #The timestamp with 3 calibration
    d3 =df[(df.gazeEngineState=='calibrating') & (df.itemID.str.contains('Calibration3'))][['PIN','Assessment Name','timestamp']].drop_duplicates()
    #The timestamp with 2 calibration
    d2 =df[(df.gazeEngineState=='calibrating')& (df.itemID.str.contains('Calibration2'))][['PIN','Assessment Name','timestamp']].drop_duplicates()
    d1 =df[(df.gazeEngineState=='calibrating') & (df.itemID.str.contains('Calibration1'))][['PIN','Assessment Name','timestamp']].drop_duplicates()
    print(d3.shape[0], 'people had 3 rounds of calibrations.' )
    print(d2.shape[0]-d3.shape[0], 'people had 2 rounds of calibrations.' )
    print(d1.shape[0]-d2.shape[0], 'people had 1 round of calibration.' )
    #13 times is the "eventName" : "failedCalibrationRound" on round 3
    third = df[(df.itemID.str.contains('Calibration3'))&(df.eventName.str.contains('failedCalibrationRound'))][['PIN','Assessment Name','timestamp']].drop_duplicates()
    print('')
    print(third.shape[0] ,'times is the "eventName" : "failedCalibrationRound" on round 3')
    display(third.head())

    two= pd.merge(d2,d3, indicator=True, how='outer').query('_merge=="left_only"').drop('_merge', axis=1)   
    one= pd.merge(d1,d2, indicator=True, how='outer').query('_merge=="left_only"').drop('_merge', axis=1)
    
    return d3,two, one, third
```

```python
def make_table(df, ff):
    failed = list(ff.PIN.unique())
    failed.append('Porter')
    sdf = df[~df.PIN.isin(failed)]
    sdf['length'] =sdf.PIN.str.len()
    sdf=sdf[sdf.length ==6]
    
    testing = sdf[sdf.gazeEngineState =='testing']
    e1 = testing.groupby(['PIN','Assessment Name','itemID'])['gazeLocationName'].value_counts(normalize=True).unstack()
    e2 =testing.groupby(['PIN','Assessment Name','itemID'])['gazeLocationOnScreen'].value_counts(normalize=True).unstack()[True]
    rv = pd.merge(e1,e2, left_index=True, right_index=True)
    rv=rv.fillna(0).reset_index()
    
    rv.columns=['PIN', 'Assessment Name', 'itemID', 'away', 'left', 'right','onscreen']
    #Four critiria
    rv.loc[(rv['onscreen'] < 0.5),'classification'] = 'missingOff'
    #rv.loc[((rv['center'] >= 0.5)&(rv['onscreen'] >= 0.5)),'classification'] = 'center'
    rv.loc[((rv['left'] >= 0.5)&(rv['onscreen'] >= 0.5)),'classification'] = 'left'
    rv.loc[((rv['right'] >= 0.5)&(rv['onscreen'] >= 0.5)),'classification'] = 'right'
    rv.loc[((rv['right'] < 0.5)&(rv['left'] < 0.5)&(rv['onscreen'] >= 0.5)),'classification'] = 'missingAway'
    
    wide =rv.pivot_table(values=['onscreen','left', 'right','classification'], index=['PIN', 'Assessment Name'], columns='itemID', aggfunc='first')
    wide.columns = [col[1]+'_'+col[0] for col in wide.columns.values]
    wide=wide.reindex(sorted(wide.columns), axis=1)
    
    return wide
```

If proportion off screen >= 50%, “missingOff”;
if proportion left > 50%, “left”;
if proportion right > 50%, “right”;
if proportion on screen > 50% but both left and right proportions are < 50%, “missingAway”
For some items on EFX, it is “center/away” instead of left/right. For those items, if proportion onscreen > 50% and proportion center >50%, then “center”, else “missingOff” or “missingAway”.
