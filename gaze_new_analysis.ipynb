# Read the json files and the registration files

```python
jfiles = glob.glob('R:\\MSS\\Research\\Projects\\Baby_Toolbox\\Gaze_Feasibility_2022_Data\\misc\\*.json')

edoc =[]
ldoc=[]
for file in jfiles:
    # load data using Python JSON module
    with open(file,'r') as f:
        ts = file.split('\\')[-1][:19]
        data = json.loads(f.read())
        print(ts)
        # Flatten data
        for i in range(len(data)):
            df0 = pd.json_normalize(data[i])
            subset =df0[(df0.dataKey=='userPIN')]
            sub =df0[(df0.dataKey=='assessmentName')]
            try:
                #display(subset.iloc[0,5])
                df0['PIN'] = subset.iloc[0,5]
            except:
                IndexError
            try:
                df0['Assessment Name'] = sub.iloc[0,5]
            except:
                IndexError
            df0['timestamp'] = ts
            #split data into LWL and EXF
            if 'LWL' in data[i][0]['itemID']:          
                ldoc.append(df0)
            else:
                edoc.append(df0)
            
            
edf = pd.concat(edoc)
ldf = pd.concat(ldoc)
```

## how many times is the "eventName" : "failedCalibrationRound" on round 3

```python
def onetwothree(df):
    #The timestamp with 3 calibration
    d3 =df[(df.gazeEngineState=='calibrating') & (df.itemID.str.contains('Calibration3'))][['PIN','Assessment Name','timestamp']].drop_duplicates()
    #The timestamp with 2 calibration
    d2 =df[(df.gazeEngineState=='calibrating')& (df.itemID.str.contains('Calibration2'))][['PIN','Assessment Name','timestamp']].drop_duplicates()
    d1 =df[(df.gazeEngineState=='calibrating') & (df.itemID.str.contains('Calibration1'))][['PIN','Assessment Name','timestamp']].drop_duplicates()
    print(d3.shape[0], 'people had 3 rounds of calibrations.' )
    print(d2.shape[0]-d3.shape[0], 'people had 2 rounds of calibrations.' )
    print(d1.shape[0]-d2.shape[0], 'people had 1 round of calibration.' )
    #13 times is the "eventName" : "failedCalibrationRound" on round 3
    third = df[(df.itemID.str.contains('Calibration3'))&(df.eventName.str.contains('failedCalibrationRound'))][['PIN','Assessment Name','timestamp']].drop_duplicates()
    print('')
    print(third.shape[0] ,'times is the "eventName" : "failedCalibrationRound" on round 3')
    display(third.head())

    two= pd.merge(d2,d3, indicator=True, how='outer').query('_merge=="left_only"').drop('_merge', axis=1)   
    one= pd.merge(d1,d2, indicator=True, how='outer').query('_merge=="left_only"').drop('_merge', axis=1)
    
    return d3,two, one, third
```

# how many times does "eventName" : "beganHeadPlacement" (regardless of gazeEngineState) per record (minimum 1)

```python
def HeadPlacement(df):
    hp = df[df.eventName=='beganHeadPlacement']
    return hp.groupby(['PIN','Assessment Name'])['eventName'].count()
```

# how many times does "eventName" : "faceLeftScene" occur (please report this separately for by gazeEngineState) per record

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('C:\\Users\\tsb7592\\Downloads\\faceLeftScene.xlsx', engine='xlsxwriter')
```

```python
def FaceLeft(df, prefix):

    fl = df[df.eventName=='faceLeftScene']
    for state in df['gazeEngineState'].unique():
        
        sub = fl[fl.gazeEngineState== state]
        if sub.shape[0]>0:
            rv =sub.groupby(['PIN','Assessment Name'])['eventName'].count()
            rv.to_excel(writer, sheet_name=prefix +state)
```
