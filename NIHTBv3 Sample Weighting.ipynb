```python
def label_race (row):
    if row['hispanic'] <= 1 :
        return 'Any Hispanic'
    elif row['race_White'] + row['race_MiddleEast'] >= 1 and row['race_AmericanIndian'] + row['race_Asian'] +row['race_Black'] + row['race_Hawaiian']+ row['race_OtherPacificIsl'] +row['race_unknwnNoAns']==0:
        return 'Non-Hispanic White'
    elif row['race_Black'] ==1  and row['race_White'] + row['race_MiddleEast'] + row['race_AmericanIndian'] + row['race_Asian'] + row['race_Hawaiian']+ row['race_OtherPacificIsl'] +row['race_unknwnNoAns']==0:
        return 'Non-Hispanic Black'
    elif  row['race_Asian'] ==1  and row['race_White'] + row['race_MiddleEast'] + row['race_AmericanIndian'] + row['race_Black'] + row['race_Hawaiian']+ row['race_OtherPacificIsl'] +row['race_unknwnNoAns']==0:
        return 'Non-Hispanic Asian'
    elif  row['race_Asian'] + row['race_White'] + row['race_MiddleEast'] + row['race_AmericanIndian'] + row['race_Black'] + row['race_Hawaiian']+ row['race_OtherPacificIsl'] +row['race_unknwnNoAns']==1:
        return 'Non-Hispanic Other Single Race'
    elif row['race_White'] + row['race_MiddleEast'] + row['race_Black'] >= 2 and row['race_AmericanIndian'] + row['race_Asian']  + row['race_Hawaiian']+ row['race_OtherPacificIsl'] +row['race_unknwnNoAns']==0:
        return 'Non-Hispanic Biracial White/Black'
    elif  row['race_White'] + row['race_MiddleEast'] + row['race_Asian'] >= 2 and row['race_AmericanIndian'] +row['race_Black'] + row['race_Hawaiian']+ row['race_OtherPacificIsl'] +row['race_unknwnNoAns']==0:
        return 'Non-Hispanic Biracial White/Asian'
    elif  row['race_Asian'] + row['race_Black'] == 2 and row['race_White'] + row['race_MiddleEast'] + row['race_AmericanIndian'] + row['race_Hawaiian']+ row['race_OtherPacificIsl'] +row['race_unknwnNoAns']==0:
        return 'Non-Hispanic Biracial Black/Asian'
    return 'Non-Hispanic Other Multiracial'    
```

```python
df['Race/ethnicity'] =df.apply (lambda row: label_race(row), axis=1)
```

```python
ad = df[df.age>=18]
kd = df[df.age<18]
```

```python
def crosstab(df):
    a =df.groupby(['region','Race/ethnicity'])['age'].count().reset_index()
    a.columns=['region','Race/ethnicity','count']
    aa =a.pivot_table(values=['count'], index=['region'], columns=['Race/ethnicity'])
    aa=aa.fillna(0)
    aa.loc['Total',:]= aa.sum(axis=0)
    aa["sum"] = aa.sum(axis=1)
    
    return aa
```
